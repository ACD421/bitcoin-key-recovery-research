```python
# Install dependencies:
# pip install pycryptodome  # For SHA256 and RIPEMD160 hashing

from Crypto.Hash import SHA256
import binascii

# Known address and your pkscript Hash160
ADDRESS_1FEEX = "1FeexV6bAHb8ybZjqQMjJrcCrHGW9sb6uF"
HASH160_FROM_PKSCRIPT = bytes.fromhex("a0b0d60e5991578ed37cbda2b17d8b2ce23ab295")

# Base58 decoding function
def base58_decode(address):
    alphabet = "123456789ABCDEFGHJKLMNPQRSTUVWXYZabcdefghijkmnopqrstuvwxyz"
    base58_map = {char: idx for idx, char in enumerate(alphabet)}
    num = 0
    for char in address:
        num = num * 58 + base58_map[char]
    byte_array = num.to_bytes((num.bit_length() + 7) // 8, "big")
    # Remove leading zeros added by encoding
    leading_zeros = len(address) - len(address.lstrip("1"))
    return b"\x00" * leading_zeros + byte_array

# Hash160 to address
def hash160_to_address(hash160):
    version = b"\x00"
    payload = version + hash160
    checksum = SHA256.new(SHA256.new(payload).digest()).digest()[:4]
    full_payload = payload + checksum
    return base58_encode(full_payload)

def base58_encode(bytes_data):
    alphabet = "123456789ABCDEFGHJKLMNPQRSTUVWXYZabcdefghijkmnopqrstuvwxyz"
    num = int.from_bytes(bytes_data, "big")
    encoded = ""
    while num > 0:
        num, remainder = divmod(num, 58)
        encoded = alphabet[remainder] + encoded
    leading_zeros = len(bytes_data) - len(bytes_data.lstrip(b"\x00"))
    return "1" * leading_zeros + encoded

# Step 1: Decode 1FeexV6 address to get its Hash160
decoded = base58_decode(ADDRESS_1FEEX)
version = decoded[0:1]
hash160_from_address = decoded[1:-4]  # Skip version byte and checksum
checksum = decoded[-4:]
computed_checksum = SHA256.new(SHA256.new(version + hash160_from_address).digest()).digest()[:4]

print("1FeexV6 Address:", ADDRESS_1FEEX)
print("Decoded Hash160 from address:", hash160_from_address.hex())
print("Checksum from address:", checksum.hex())
print("Computed checksum:", computed_checksum.hex())
print("Checksum match:", checksum == computed_checksum)

# Step 2: Compare with your pkscript Hash160
print("\nYour pkscript Hash160:", HASH160_FROM_PKSCRIPT.hex())
print("Matches 1FeexV6 Hash160?", hash160_from_address == HASH160_FROM_PKSCRIPT)

# Step 3: Convert your pkscript Hash160 to an address
address_from_pkscript = hash160_to_address(HASH160_FROM_PKSCRIPT)
print("Address from your pkscript Hash160:", address_from_pkscript)
print("Matches 1FeexV6?", address_from_pkscript == ADDRESS_1FEEX)
```

```python
#!/usr/bin/env python3

import base58
import binascii

def verify_hash160(address: str, expected_hex: str) -> bool:
    # Decode Base58Check
    payload_bytes = base58.b58decode(address)
    # Convert to hex for easy slicing
    payload_hex = binascii.hexlify(payload_bytes).decode()

    # Typical P2PKH format:
    #   [1-byte version] + [20-byte HASH160] + [4-byte checksum]
    version_byte = payload_hex[0:2]        # '00' for mainnet P2PKH
    hash160_hex  = payload_hex[2:-8]       # 20 bytes => 40 hex chars
    checksum_hex = payload_hex[-8:]        # 4 bytes => 8 hex chars

    print(f"Address:          {address}")
    print(f"Decoded Hex:      {payload_hex}")
    print(f"Version Byte:     {version_byte}")
    print(f"Hash160 (hex):    {hash160_hex}")
    print(f"Checksum (hex):   {checksum_hex}")

    # Compare to expected
    return (hash160_hex.lower() == expected_hex.lower())

if __name__ == "__main__":
    addr = "1FeexV6bAHb8ybZjqQMjJrcCrHGW9sb6uF"
    known_hash160 = "a0b0d60e5991578ed37cbda2b17d8b2ce23ab295"

    print("\nVerifying Hash160 for address:", addr)
    result = verify_hash160(addr, known_hash160)
    if result:
        print(f"\nSUCCESS: Hash160 matches {known_hash160}")
    else:
        print("\nFAILURE: Hash160 did NOT match expected value!")
```

```python
#!/usr/bin/env python3

import binascii
import base58
import hashlib

# pkscript:
# 76 a9 14 a0 b0 d6 0e 59 91 57 8e d3 7c bd a2 b1 7d 8b 2c e2 3a b2 95 88 ac
# OP_DUP OP_HASH160 <20-byte-hash> OP_EQUALVERIFY OP_CHECKSIG

PKSCRIPT_HEX = "76a914a0b0d60e5991578ed37cbda2b17d8b2ce23ab29588ac"
TARGET_ADDRESS = "1FeexV6bAHb8ybZjqQMjJrcCrHGW9sb6uF"

# 1) Extract the 20-byte hash160 from pkscript:
#    pkscript format: OP_DUP (0x76), OP_HASH160 (0xa9), push-20 (0x14),
#    then <20-byte hash160>, then OP_EQUALVERIFY (0x88), OP_CHECKSIG (0xac).

# The 20 bytes (40 hex chars) start right after "76 a9 14":
HASH160_HEX = PKSCRIPT_HEX[6:6+40]  # skip "76a914" => next 40 hex chars
print(f"Extracted Hash160 from pkscript: {HASH160_HEX}")

# 2) Prepend version byte "00"
payload_hex = "00" + HASH160_HEX

# 3) Double-SHA256 the 21-byte payload, take first 4 bytes as checksum
payload_bytes = binascii.unhexlify(payload_hex)
checksum_full = hashlib.sha256(hashlib.sha256(payload_bytes).digest()).digest()
checksum_4 = checksum_full[:4]
checksum_hex = binascii.hexlify(checksum_4).decode()
print(f"Computed 4-byte checksum: {checksum_hex}")

# 4) Construct the final 25-byte data => Base58Check encode => address
full_hex = payload_hex + checksum_hex
full_bytes = binascii.unhexlify(full_hex)
calc_address = base58.b58encode(full_bytes).decode()

print(f"Calculated Address: {calc_address}")

# 5) Compare
if calc_address == TARGET_ADDRESS:
    print(f"\nSUCCESS: pkscript {PKSCRIPT_HEX} indeed matches address {TARGET_ADDRESS}.")
else:
    print(f"\nFAILURE: mismatch. Computed {calc_address} != {TARGET_ADDRESS}")
```
